### <div class="alert alert-block alert-info"><b>Import library<b><div>


```python
# Import required libraries
# pandas → data manipulation
# numpy → numerical operations

import pandas as pd
import numpy as np

# Display full columns
pd.set_option('display.max_columns', None)
```

### <div class="alert alert-block alert-info"><b>Load dataset<b><div>


```python
# Load dataset
# Load match-level dataset
matches=pd.read_csv("../Datasets/matches.csv")
# Load ball-by-ball dataset
deliveries=pd.read_csv("../Datasets/deliveries.csv")
```

### <div class="alert alert-block alert-info"><b>Data Validation <b><div>

```python
deliveries.info()
print(30*"==")
deliveries.isnull().sum()
```

### <div class="alert alert-block alert-info"><b>Remove Duplicate Rows<b><div>


```python
# Remove duplicate rows from datasets
matches.drop_duplicates(inplace=True)
deliveries.drop_duplicates(inplace=True)

print("Duplicates removed successfully")
```

### <div class="alert alert-block alert-info"><b>Handle Missing Values (matches)<b><div>


```python
# Fill missing city values using venue

matches["city"] = matches["city"].fillna("Unknown")
```

```python
# Fill missing winner value

matches["winner"]=matches["winner"].fillna("No Result")
```

```python
matches["player_of_match"]=matches["player_of_match"].fillna("Not Awarded")
```

```python
# Handle result_margin
# NULL → 0
matches["result_margin"] = matches["result_margin"].fillna(0)
```

```python
matches["target_runs"]=matches["target_runs"].fillna(0)

matches["target_overs"]=matches["target_overs"].fillna(0)
```

```python
# Handle method column
# NULL → Normal match
matches["method"] = matches["method"].fillna("Normal")
```

### <div class="alert alert-block alert-info"><b>City Name Standardization <b><div>


```python
#city name Standardize
matches["city"]=matches["city"].replace({ "Bangalore" :"Bengaluru"})
# verify
matches[matches["city"]=="Bengalore" ]
```

### <div class="alert alert-block alert-info"><b>Team Name Standardization Both dataset<b><div>


```python
# Standardize Team Names
Team_standardization= {"Delhi Daredevils":"Delhi Capitals",
                      "Kings XI Punjab":"Punjab Kings" ,
                      "Rising Pune Supergiant" :"Rising Pune Supergiants" ,
                      "Royal Challengers Bangalore":"Royal Challengers Bengaluru" }

```

```python
#Team - matches dataset

Team_columns=["team1","team2","toss_winner","winner"]
matches[Team_columns]=matches[Team_columns].replace(Team_standardization)
# verify
matches[Team_columns]
```

```python
# Name standardization (batting_team , bowling_team) in deliveries dataset 

bat_bowl_column=["batting_team","bowling_team"]
deliveries[bat_bowl_column]=deliveries[bat_bowl_column].replace(Team_standardization)

#verify
deliveries[bat_bowl_column]
```

### <div class="alert alert-block alert-info"><b>Convert datatype <b><div>

```python
# To Ensure Numeric Columns Proper
numeric_cols = [
    
    "batsman_runs",
    "extra_runs",
    "total_runs",
    "is_wicket"
]

deliveries[numeric_cols] = deliveries[numeric_cols].apply(pd.to_numeric)
```

```python
# Convert date column to datetime format
# Dataset date format → DD-MM-YYYY
matches["date"] = pd.to_datetime(matches["date"],format="%d-%m-%Y")
print(matches["date"])
```

### <div class="alert alert-block alert-info"><b>Create Column<b><div>


```python
#year column create
matches["year"]=matches["date"].dt.year
```

```python
# Create match_phase column
def match_phase(over):
    
    if over <= 6:
        return "Powerplay"
    
    elif over <= 15:
        return "Middle Overs"
    
    else:
        return "Death Overs"

deliveries["match_phase"] = deliveries[
    "over"
].apply(match_phase)
```

```python
# Create super_over flag
 #HANDLE INNING VALUES

deliveries["is_super_over"] = deliveries["inning"].apply(
    lambda x: 1 if x > 2 else 0
)
```

### <div class="alert alert-block alert-info"><b>Handle Missing Values (deliveries)<b><div>


```python
# Fill missing extras_type
#  NULL → No Extra
deliveries["extras_type"] = deliveries["extras_type"].fillna("No Extra")
```

```python
# player dismissed - NULL → Not Out
deliveries["player_dismissed"]=deliveries["player_dismissed"].fillna("Not Out")

# fielder NULL → Not out
deliveries["dismissal_kind"]=deliveries["dismissal_kind"].fillna("Not Out")

# fielder NULL
deliveries["fielder"]=deliveries["fielder"].fillna("None")
```

```python
# Remove invalid overs if present more than 20 .
deliveries = deliveries[deliveries["over"] <= 20]
```

```python
deliveries = deliveries[deliveries["ball"] <= 6]
```

### <div class="alert alert-block alert-info"><b>Merge Datasets<b><div>


```python
ipl_data = pd.merge(deliveries , matches , on="match_id" , how="left" )
```
